```python
# Import Library
import pandas as pd
import numpy as np
from tqdm.auto import tqdm
import time
```

```python
class color:
    PURPLE = '\033[95m'
    CYAN = '\033[96m'
    DARKCYAN = '\033[36m'
    BLUE = '\033[94m'
    GREEN = '\033[92m'
    YELLOW = '\033[93m'
    RED = '\033[91m'
    BOLD = '\033[1m'
    UNDERLINE = '\033[4m'
    END = '\033[0m'
```

```python
tanggal = []
i = 5
while i < 30:
    a = "%02.f" % i
    tanggal.append(f"2023-12-{a}")
    # print(a)
    i+=1

# print(tanggal)
```

```python
def func1(date):
    # Import Dataset
    # date = "2023-12-29"
    df_pres = pd.read_csv(f'C:/Users/labib/Documents/code/Skripsi/Tugas-Akhir-P24/anies/Anies_{date}.csv', lineterminator='\n')

    df_wpres = pd.read_csv(f'C:/Users/labib/Documents/code/Skripsi/Tugas-Akhir-P24/anies/Muhaimin_{date}.csv', lineterminator='\n')

    df_acr = pd.read_csv(f'C:/Users/labib/Documents/code/Skripsi/Tugas-Akhir-P24/anies/Amin_{date}.csv', lineterminator='\n')

    # count total data
    print("Pres:", len(df_pres), " Wapres:", len(df_wpres), " Akronim:", len(df_acr))
    total_data = len(df_pres) + len(df_wpres) + len(df_acr)
    print("Total Data: ", total_data)

    # merge dataset
    df_pres['keyword'] = 'Pres'
    df_wpres['keyword'] = 'Wapres'
    df_acr['keyword'] = 'Akronim'
    df = pd.concat([df_pres, df_wpres, df_acr], axis=0, ignore_index=True)
    
    # count duplicate data
    print("Duplicate Data: ", df.duplicated().sum())

    # count duplicate data based on full_text
    print("Duplicate Data by Text: ", df['full_text'].duplicated().sum())

    # drop duplicate data based on full_text
    df = df.drop_duplicates(subset=['full_text'])
    print("Total Data after Drop: ", len(df))

    df = df.sort_values(by=['created_at'])
    
    df_len = len(df)
    print(f"Total Data: {df_len}")

    df.to_csv(f'./cleaned_dataset/paslon1/Amin_{date}.csv', index=False)
    # time.sleep(0.5)
```

```python
for i in tqdm(tanggal):
    print(f"\n***********\n{color.BOLD+color.GREEN}Tanggal: {i}{color.END}")
    func1(i)
```
