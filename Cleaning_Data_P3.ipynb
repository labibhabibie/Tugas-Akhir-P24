```python
# Import Library
import pandas as pd
import numpy as np
```

```python
# Import Dataset
date = "2023-12-29"
df_pres1 = pd.read_csv(f'./ganjar/Ganjar_{date}.csv', lineterminator='\n')
# df_pres2 = pd.read_csv(f'./ganjar/2_Prabowo_{date}.csv', lineterminator='\n')
# df_pres = pd.concat([df_pres1, df_pres2], axis=0, ignore_index=True)
# df_pres = df_pres.drop_duplicates(subset=['full_text'])

df_wpres = pd.read_csv(f'./ganjar/GaMa_{date}.csv', lineterminator='\n')
df_acr = pd.read_csv(f'./ganjar/Mahfud_{date}.csv', lineterminator='\n')
```

```python
# count total data
print("Pres:", len(df_pres1), " Wapres:", len(df_wpres), " Akronim:", len(df_acr))
total_data = len(df_pres1) + len(df_wpres) + len(df_acr)
print("Total Data: ", total_data)
```

```python
# merge dataset
df_pres1['keyword'] = 'Pres'
df_wpres['keyword'] = 'Wapres'
df_acr['keyword'] = 'Akronim'
df = pd.concat([df_pres1, df_wpres, df_acr], axis=0, ignore_index=True)
df
```

```python
# count duplicate data
print("Duplicate Data: ", df.duplicated().sum())
```

```python
# count duplicate data based on full_text
print("Duplicate Data by Text: ", df['full_text'].duplicated().sum())
```

```python
# drop duplicate data based on full_text
df = df.drop_duplicates(subset=['full_text'])
print("Total Data after Drop: ", len(df))
```

```python
# sort data by created_at
df = df.sort_values(by=['created_at'])
df
```
