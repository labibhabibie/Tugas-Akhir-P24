```python
# Import Library
import pandas as pd
import numpy as np
```

```python
# Import Dataset
date = "2023-12-05"
df_pres = pd.read_csv(f'./Prabowo/Prabowo_{date}.csv')
df_wpres = pd.read_csv(f'./Prabowo/Gibran_{date}.csv')
df_acr = pd.read_csv(f'./Prabowo/PraGi_{date}.csv')
```

```python
# count total data
print("Pres:", len(df_pres), " Wapres:", len(df_wpres), " Akronim:", len(df_acr))
total_data = len(df_pres) + len(df_wpres) + len(df_acr)
print("Total Data: ", total_data)
```

```python
# merge dataset
df_pres['keyword'] = 'Pres'
df_wpres['keyword'] = 'Wapres'
df_acr['keyword'] = 'Akronim'
df = pd.concat([df_pres, df_wpres, df_acr], axis=0, ignore_index=True)
df
```

```python
# count duplicate data
print("Duplicate Data: ", df.duplicated().sum())
```

```python
# count duplicate data based on full_text
print("Duplicate Data by Text: ", df['full_text'].duplicated().sum())
```

```python
# drop duplicate data based on full_text
df = df.drop_duplicates(subset=['full_text'])
print("Total Data after Drop: ", len(df))
```

```python
# take 250 random data but include keyword Akronim
df = df.sample(n=250, random_state=42)
df = df.sort_values(by=['created_at'])
df
```
