```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Membaca dataset dari file CSV (gantilah 'nama_file.csv' dengan nama file dataset Anda)
df = pd.read_csv('dataset13-22.csv')

# Menghitung total retweets, replies, dan likes untuk setiap user
user_popularity = df.groupby('username').agg({
    'retweets_count': 'sum',
    'replies_count': 'sum',
    'likes_count': 'sum'
}).reset_index()

# Menghitung total tweet untuk setiap user
user_popularity['total_tweets'] = df['username'].value_counts().reset_index()['username']

# Menghitung popularitas dengan menggunakan metrik tertentu, misalnya jumlah retweets + replies + likes
user_popularity['popularity_score'] = user_popularity['retweets_count'] + user_popularity['replies_count'] + user_popularity['likes_count']

# Mengurutkan hasil berdasarkan popularitas score
user_popularity = user_popularity.sort_values(by='popularity_score', ascending=False)

# Menampilkan 10 user teratas berdasarkan popularitas
top_10_users = user_popularity.head(10)

# Visualisasi data
plt.figure(figsize=(12, 6))
sns.barplot(x='popularity_score', y='username', data=top_10_users, palette='viridis')
plt.xlabel('Popularitas Score (Retweets + Replies + Likes)')
plt.ylabel('Username')
plt.title('Top 10 Most Popular Twitter Users')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Membaca dataset dari file CSV (gantilah 'nama_file.csv' dengan nama file dataset Anda)
df = pd.read_csv('dataset13-22.csv')

# Menghitung seberapa sering setiap pengguna disebutkan (mention) dalam tweet
mention_count = df['mentions'].str.split(', ').explode().value_counts().reset_index()
mention_count.columns = ['username', 'mention_count']

# Mengurutkan hasil berdasarkan jumlah mention
mention_count = mention_count.sort_values(by='mention_count', ascending=False)

# Visualisasi data
plt.figure(figsize=(12, 6))
sns.barplot(x='mention_count', y='username', data=mention_count.head(10), palette='viridis')
plt.xlabel('Jumlah Mention')
plt.ylabel('Username')
plt.title('Top 10 Most Mentioned Twitter Users')
plt.show()
```

```python
import pandas as pd
# Load the data
df = pd.read_csv('dataset13-22.csv')
# Create a new column to store the popularity of each user
df['popularity'] = 0
# Iterate over each row in the data
for index, row in df.iterrows():
    # Get the number of replies, retweets, and likes for the user
    replies_count = row['replies_count']
    retweets_count = row['retweets_count']
    likes_count = row['likes_count']
    # Calculate the popularity of the user
    popularity = replies_count + retweets_count + likes_count
    # Update the popularity column for the user
    df.at[index, 'popularity'] = popularity
# Print the popularity of each user
print(df['popularity'])
```
